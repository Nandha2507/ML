### 8) Dropping the ID feature 

```python
df.drop(["Id"],axis = 1,inplace = True)
df.head()
```

### 9) Splitting the data into Independent and dependent features

```python
feature_cols = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
X = df[feature_cols]
y = df.Species
```

### 10) Viewing X and y

```python
print(X)
```

```python
print(y)
```

###  11)Implementing Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

### 12) Import the DecisionTreeClassifier form Sklearn and fit the model

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'gini', random_state = 0)
classifier.fit(X_train, y_train)
```

### 13) Predict the test data and print the same

```python
y_pred = classifier.predict(X_test)
print(y_pred)
```

### 14) From Sklearn Import Accuracy score and check the accuracy of the model

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred)*100 ,'%')
```

### 15) To visualize the Decision Tree import the following libraries

```python
import pydotplus
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO  
from IPython.display import Image
from sklearn import tree
import graphviz
```

### 16) Visualize the Decision Tree

```python
plt.figure(figsize=(24,15))
tree.plot_tree(classifier.fit(X,y),feature_names=feature_cols,filled=True,precision=3,
              proportion=True,rounded=True)
plt.show()
```

### 17) From Sklearn import Confusion_matrix and print the confusion matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```
